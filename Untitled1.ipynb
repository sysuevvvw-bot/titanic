```python
import ast
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import tree
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
%matplotlib inline
```

```python
events_data = pd.read_csv("event_data_train.csv")
submissions_data = pd.read_csv("submissions_data_train.csv")
```

```python
#перевод дат
events_data['data'] = pd.to_datetime(events_data.timestamp, unit = 's')
submissions_data['data'] = pd.to_datetime(submissions_data.timestamp, unit = 's')
events_data['day'] = events_data.data.dt.date
submissions_data['day'] = submissions_data.data.dt.date
```

```python
#создание таблицы со статусом задач
users_events_data = events_data.pivot_table(index = 'user_id',
                                            columns = 'action',
                                            values = 'step_id',
                                            aggfunc = 'count',
                                            fill_value = 0).reset_index()
```

```python
#создание таблицы с кол-вом верных и неверных попыток
users_scores = submissions_data.pivot_table(index = 'user_id',
                                            columns = 'submission_status',
                                            values = 'step_id',
                                            aggfunc = 'count',
                                            fill_value = 0).reset_index()
```

```python
users_data = events_data.groupby('user_id', as_index = False).\
    agg({'timestamp': 'max'}).\
    rename(columns = {'timestamp' : 'last_timestamp'})
#максимальное время бездействия поользователя
```

```python
now = 1526772811 #дата сейчас
drop_out = 2592000 #30 дней
```

```python
users_data['on_users'] = (now - users_data.last_timestamp) > drop_out  # True = дропнутый пользователь
```

```python
#количество дней пользователя на степике
users_days = events_data.groupby('user_id').day.nunique()
users_days = users_days.to_frame().reset_index()
```

```python
#объединение датафреймов
users_data = users_data.merge(users_scores, on = 'user_id', how = 'outer')
users_data = users_data.merge(users_events_data, how = 'outer')
users_data = users_data.merge(users_days, how = 'outer')
```

```python
#определение пользователей прошедших курс
users_data['passed_course'] = users_data.passed > 170
```

```python
user_min_time = events_data.groupby('user_id', as_index = False) \
.agg({'timestamp': 'min'}) \
.rename({'timestamp': 'min_timestamp'}, axis = 1)
```

```python
users_data = users_data.merge(user_min_time, how='outer')
```

Сделать для каждого юзера промежуток три дня от начала прохождения курса

```python
learning_time_threshold = 30*24*60*60
```

```python
user_min_time = events_data.groupby('user_id', as_index=False) \
    .agg({'timestamp': 'min'}) \
    .rename(columns={'timestamp': 'min_timestamp'})
```

```python
events_data = events_data.merge(user_min_time, on='user_id', how='left')
submissions_data = submissions_data.merge(user_min_time, on='user_id', how='left')
```

```python
events_data_train = events_data[events_data.timestamp <= events_data.min_timestamp + learning_time_threshold]
events_data_train.groupby('user_id').day.nunique().max()
```

```python
submissions_data_train = submissions_data[submissions_data.timestamp <= submissions_data.min_timestamp + learning_time_threshold]
submissions_data_train.groupby('user_id').day.nunique().max()
```

```python
X = submissions_data_train.groupby('user_id').day.nunique().to_frame().reset_index().rename(columns={'day': 'days'})
```

```python
steps_tried = submissions_data_train.groupby('user_id').step_id.nunique().to_frame().reset_index().rename(columns = {'step_id': 'steps_tried'})
```

```python
X = X.merge(steps_tried, on = 'user_id', how = 'outer')
```

```python
X = X.merge(submissions_data_train.pivot_table(index = 'user_id',
                                            columns = 'submission_status',
                                            values = 'step_id',
                                            aggfunc = 'count',
                                            fill_value = 0).reset_index())
```

```python
X['correct_ratio'] = X['correct'] / (X['correct'] + X['wrong'])
```

```python
X = X.merge(events_data_train.pivot_table(index = 'user_id',
                                            columns = 'action',
                                            values = 'step_id',
                                            aggfunc = 'count',
                                            fill_value = 0).reset_index()[['user_id', 'viewed']], how = 'outer')
```

```python
X = X.fillna(0)
```

```python
X = X.merge(users_data[['user_id', 'on_users', 'passed_course']], how = 'left')
```

```python
X = X[~((X.passed_course == False) & (X.on_users == False))]
```

```python
y = X.passed_course.map(int)
```

```python
X = X.set_index(X.user_id)
X = X.drop(['passed_course', 'on_users', 'user_id'], axis = 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 42, stratify = y)
```

```python
parameters = {'criterion' : ['gini', 'entropy'], 'max_depth' : range(1,5)}
clf = tree.DecisionTreeClassifier(class_weight='balanced')
grid_search_cv_clf = GridSearchCV(clf, parameters, cv = 5)
```

```python
grid_search_cv_clf.fit(X_train, y_train)
best_clf = grid_search_cv_clf.best_estimator_
```

```python
best_clf.score(X_test, y_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

y_pred = best_clf.predict(X_test)
cm = confusion_matrix(y_test, y_pred)

# Визуализация
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
from sklearn.metrics import roc_auc_score

# Получаем вероятности для положительного класса (индекс 1)
y_predicted_prob = best_clf.predict_proba(X_test)[:, 1]
print(f'ROC-AUC score: {roc_auc_score(y_test, y_predicted_prob)}')
```

```python
print(f"Best CV score: {grid_search_cv_clf.best_score_}")
```
